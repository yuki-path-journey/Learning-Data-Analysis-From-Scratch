```python
plt.hist(train['SalePrice'], color='lightblue', edgecolor='white')
plt.title('Housing Prices Distribution')
plt.xlabel('Sale Price ($)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(train['SalePrice'].mean(), color='red', linestyle='-', linewidth=2)
plt.show()
```

「基本のヒストグラムに color 引数を追加して、lightblue（水色）にカスタマイズした。引数によるグラフの調整方法を学んだ。」

タイトルをつける（作品にする）

横軸の Sale Price ($)と縦軸の Frequencyを追加

水色の棒グラフに、**白い（white）の縁取り（edgecolor）**をつけて、1本ずつの区切りをはっきりさせた。

「**y軸（縦方向）**の目盛りに合わせて、**点線（'--'）のガイドを引きなさい。ただし、グラフの邪魔をしないように少し透かして（alpha=0.7）**ね」

**1.**    　1枚目: 基本の形（タイトルと横軸）

**2.**   　2枚目: 縦軸の「名札（Frequency）」がついて、図解として親切に

**3.**   　3枚目: 白い縁取り（edgecolor）で、データの塊がパキッと独立

**4.**  　 4枚目（最新）: 補助線（grid）が入り、**「正確に読み取れるグラフ」**へ

「**垂直な線（axvline）を引きなさい。場所は SalePrice の平均値（mean）のところで。色は目立つように赤（red）**にしてね」

これを足すと、グラフの山に対して、平均的な家の価格がどこにあるのかが一目でわかります。

**３**:　　「はっきりと散っている」部分（右側の端っこ）

ここがまさに**「個性が強い家」**たちです。「広くて超高い家」もあれば、さっき見つけた「広いのに安い家（外れ値）」もあります。

```python
import seaborn as sns

sns.boxplot(x='OverallQual', y='SalePrice', data=train)
plt.title('Quality vs Sale Price')
plt.show()
```

このコードでわかること：

箱（Box）: その評価の中で、真ん中くらいの価格帯がどこか。

ひげ（Whisker）: 価格がどこまで伸びているか。

箱ひげ図を見るときに必ず出てくる言葉です。

**Median（メディアン / 中央値）**

意味: 箱の中にある「横線」のこと。平均値（Mean）と違って、極端な豪邸に振り回されない「ちょうど真ん中の順位」の価格です。

**Interquartile Range（IQR / 四分位範囲）**

意味: **「箱そのものの長さ」**のこと。この箱の中に、全データの真ん中50%が収まっています。箱が長いほど、同じ評価でも「価格にバラツキがある」ということがわかります。

**1：** 綺麗な「階段」になっている

品質（OverallQual）が1から10に上がるにつれて、箱の位置が面白いほどきれいに上がっていますね。これは、**「品質スコアは価格を予測する上で最強の武器になる」**ということを示しています。

**2：** 品質が高くなるほど「ばらつき」が激しい

評価5くらいまでは箱が小さいですが、評価10になると箱がぐーんと縦に長くなっています。

これは「最高品質の家の中でも、ものすごく高い家と、そこそこの価格の家で差が激しい」ということを表しています。

**3**：箱の外の「○（まる）」たち

箱の上下にある「○」は、その品質グループの中でも例外的な価格の家です。これらもまた**「外れ値」**の候補ですね。

```python
sns.boxplot(x='OverallQual', y='SalePrice', data=train, palette='viridis')
```

palette='viridis': これを足すと、品質が上がるにつれて色が「紫→緑→黄色」と変化する美しいグラデーションになります。

```python
import seaborn as sns

sns.scatterplot(x='GrLivArea', y='SalePrice', hue='OverallQual', palette='viridis', data=train)
plt.title('Area & Quality vs Sale Price')
plt.show()
```

横軸: 家の広さ

縦軸: 家の価格

**色 (hue)**: 家の品質（1〜10）

これを実行すると、**「広いけれど、品質が低いから安い家」や「狭いけれど、品質が最高だから高い家」**が、色の違いで一目でわかるようになります。

**1**：**左下から右上への「色のグラデーション」**

左下（狭くて安い）は濃い紫。

右上（広くて高い）は鮮やかな黄色。

きれいに色が層になって分かれていますね。これは「広さと品質がセットで価格を押し上げている」証拠です。

**2**：**右下の「怪物」たちの正体**

面積が5000を超える一番右側の点を見てください。色は**「黄色（品質10）」**ですね！

「最高品質で、最大級の広さ」なのに、価格は真ん中くらい（20万ドル付近）に沈んでいます。
